## 2. Data Loading and Cleaning
- Load CSV with `pandas.read_csv()`
- Check missing values, data types



```python
import pandas as pd
import numpy as np
import random
import datetime
import json

CATEGORIES_JSON = 'US_category_id.json'
RAW_DATA_CSV = './CAvideos.csv'

#Load YouTube raw data
data = pd.read_csv(RAW_DATA_CSV)
data.isnull().sum()
print('\n\ndata.info():')
print(data.info())
```

```python
# change the dates into datetime objects
data['trending_date'] = pd.to_datetime(data['trending_date'], format = "%y.%d.%m")
data['publish_time'] = pd.to_datetime( data['publish_time'], format = "%Y-%m-%dT%H:%M:%S.%fZ")

# Load category mapping just once
with open(CATEGORIES_JSON, 'r') as f:
    categories = json.load(f)

# Build mapping dictionary
id_to_category = {int(cat['id']): cat['snippet']['title'] for cat in categories['items']}

# Apply mapping to DataFrame
data['category_id'] = data['category_id'].map(id_to_category).astype(str)

#Remove all columns with text to make it easy for tableau
no_text =  data.drop(labels = ['video_id', 'category_id', 'title', 'channel_title', 'tags', 'thumbnail_link','description'], axis=1, inplace=False)
print('\n\nno_text.info():')
print(no_text.info())

no_text.to_csv('YT_NO_TEXT.csv')
```

## 3. Exploratory Data Analysis
- Top categories
- Likes-to-views ratio
- Best publish time
- Correlation heatmap


```python
import matplotlib.pyplot as plt

#Check top categories
top_categories = data['category_id'].value_counts()

# Plot top 10 categories
top_categories[:10].plot(kind='bar', figsize=(10,6), color='skyblue')

plt.title("Top 10 YouTube Trending Categories")
plt.xlabel("Category")
plt.ylabel("Number of Trending Videos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import textwrap

# Plot all category distributions
# Count the frequency of each category_id
category_counts = data['category_id'].value_counts()

# Get top 10 category_ids
top_10 = category_counts.head(10)
other = category_counts[10:]

# Combine "Other" category
top_10['Other'] = other.sum()

# Prepare labels (category_id for now; can map to names later)
labels = top_10.index.astype(str)

# Plot pie chart
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(top_10, labels=labels, autopct='%1.1f%%', startangle=140)

# Add legend for "Other"
if 'Other' in labels:
    plt.legend(wedges, labels, title="Category ID", loc="center left", bbox_to_anchor=(1.3, 0.8))
    other_ids = other.index.astype(str).tolist()
    other_text = "Other includes:\n" + ", ".join(other_ids)
    wrapped_text = "\n".join(textwrap.wrap(other_text, width=40))  # adjust width as needed
    plt.text(2, 0.2, wrapped_text, fontsize=9, verticalalignment='top')

    
plt.title("YouTube Video Category Distribution (Top 10 and Other)")
plt.tight_layout()
plt.show()

```

```python
# Convert publish_time to datetime and extract the hour
data['publish_time'] = pd.to_datetime(data['publish_time'])
data['publish_hour'] = data['publish_time'].dt.hour

# Calculate views per hour (total)
hourly_views = data.groupby('publish_hour')['views'].sum()

# Calculate publish count per hour
publish_counts = data.groupby('publish_hour')['views'].count()

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart: Number of videos published
ax1.bar(publish_counts.index, publish_counts.values, color='green', alpha=0.8, label='Video Count')
ax1.set_xlabel('Hour of Day (0â€“23)')
ax1.set_ylabel('Number of Videos Published', color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Line chart: Total views by publish hour
ax2 = ax1.twinx()
ax2.plot(hourly_views.index, hourly_views.values, color='blue', marker='o', label='Total Views')
ax2.set_ylabel('Total Views', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Title and layout
plt.title('Publish Time vs. View Count by Hour')
fig.tight_layout()
plt.grid(axis='x')
plt.show()
```

```python

# Group by category_id and calculate average metrics
category_stats = data.groupby('category_id')[['likes', 'dislikes', 'comment_count']].mean()

# Sort by category_id for consistent order
category_stats = category_stats.sort_index()

# Create x-axis positions
x = np.arange(len(category_stats))
labels = category_stats.index.astype(str)

# Start figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))

#show bars for average comment count
bars = ax1.bar(x, category_stats['comment_count'], color='lightgrey', alpha=0.5, label='Avg Comments')
ax1.set_ylabel('Avg Comments', color='gray')
ax1.tick_params(axis='y', labelcolor='gray')

# Create second y-axis for line charts
ax2 = ax1.twinx()

# Plot likes/dislikes/comments as line charts
ax2.plot(x, category_stats['likes'], color='blue', marker='o', label='Avg Likes')
ax2.plot(x, category_stats['dislikes'], color='orange', marker='o', label='Avg Dislikes')


ax2.set_ylabel('Avg Likes / Dislikes', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# X-axis labels
ax1.set_xticks(x)
ax1.set_xticklabels(labels, rotation=75)
ax1.set_xlabel('Category ID')

# Title and legend
plt.title("YouTube Video Engagement by Category (Average)")
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

## 4. Statistical Insights
- Confidence intervals on engagement metrics
- Viewer behavior patterns


```python
import numpy as np
from scipy import stats

music_data = data[data['category_id'] == 'Music']
ratios = music_data['likes'] / music_data['views']

# Compute mean and 95% confidence interval
mean_ratio = ratios.mean()
sem = stats.sem(ratios)  # standard error of the mean
ci = stats.t.interval(0.95, len(ratios)-1, loc=mean_ratio, scale=sem)

lower, upper = ci
print(f"Mean like/view ratio: {mean_ratio*100:.4f}%")
print(f"95% confidence interval: ({lower*100:.4f}%, {upper*100:.4f}%)")
```

```python
from scipy.stats import ttest_ind

#Find the comment rate
pets = data[data['category_id'] == 'Pets & Animals'].copy()
sports = data[data['category_id'] == 'Sports'].copy()
pets['comment_rate'] = pets['comment_count'] / pets['views'] * 1000
sports['comment_rate'] = sports['comment_count'] / sports['views'] * 1000
print("Pets & Animals mean comment rate:", pets['comment_rate'].mean())
print("Sports mean comment rate:", sports['comment_rate'].mean())

#Run the T-test
t_stat, p_value = ttest_ind(pets['comment_rate'].dropna(), sports['comment_rate'].dropna(), equal_var=False)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```
